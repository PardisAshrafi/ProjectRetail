```python
df = pd.read_excel(r"C:\Users\pardis\Desktop\ProjectRetail\Online Retail.xlsx")
df.head(3)
df.shape
```

```python
df.info()
```

```python
df = df.dropna(subset=["CustomerID"])
```

```python
df = df[df["Quantity"] > 0]
```

```python
df = df[df["UnitPrice"] > 0]
```

```python
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["OrderDate"] = df["InvoiceDate"].dt.date
df["OrderMonth"] = df["InvoiceDate"].dt.to_period("M").astype(str)
```
